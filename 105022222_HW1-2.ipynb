```
# First kind of Randam Walk
import random 
import numpy as np 
import matplotlib.pyplot as plt 
import math
step_n = 10000
path_x = np.random.uniform(-1/2, 1/2, step_n)
path_x = np.cumsum(path_x)
path_y = np.random.uniform(-1/2, 1/2, step_n)
path_y = np.cumsum(path_y)

plt.plot(path_x, path_y)
```

```
x_path_end_point = []
y_path_end_point = []
for i in range(10000):
    step_number = 10000
    path_x = np.random.uniform(-1/2, 1/2, step_number)
    x_path_end_point.append(np.sum(path_x))
    path_y = np.random.uniform(-1/2, 1/2, step_number)
    y_path_end_point.append(np.sum(path_y))
    
plt.plot(x_path_end_point, y_path_end_point,'.')
```

```
path_end_point = []

for i in range(10000):  
    step_number = 10000 
    path_x = np.random.uniform(-1/2, 1/2, step_number)
    x = np.sum(path_x)
    path_y = np.random.uniform(-1/2, 1/2, step_number)
    y = np.sum(path_y)
    path_end_point.append(math.sqrt((x**2)+(y**2)))
    
plt.hist(path_end_point)
```

```
# Second kind of Randam Walk
step_number = 10000
step_set = [-1, 1]
origin = np.zeros((1,2))
step_shape = (step_number,2)
steps = np.random.choice(a=step_set, size=step_shape, p=[0.5, 0.5])
path = np.concatenate([origin, steps]).cumsum(0)

plt.plot(path[:,0],path[:,1])
```

```
x_path_end_point = []
y_path_end_point = []
for i in range(10000):
    step_number = 10000
    step_set = [-1, 1]
    origin = np.zeros((1,2))
    step_shape = (step_number,2)
    steps = np.random.choice(a=step_set, size=step_shape, p=[0.5, 0.5])
    path = np.concatenate([origin, steps]).cumsum(0)
    x_path_end_point.append(path[-1][0])
    y_path_end_point.append(path[-1][1])
    
plt.plot(x_path_end_point, y_path_end_point,'.')
```
