```
# First kind of Randam Walk
import random 
import numpy as np 
import matplotlib.pyplot as plt 

step_number = 10000
path = np.random.uniform(-1/2, 1/2, step_number)
path = np.cumsum(path)
plt.title("Single Trajectory")
plt.xlabel('Step')
plt.ylabel('Location')
plt.plot(path)
```

```
path_end_point = []
for i in range(10000): 
    step_number = 10000  
    path = np.random.uniform(-1/2, 1/2, step_number)
    path_end_point.append(np.sum(path))

plt.title("End point of all Random walk")
plt.xlabel('Episode')
plt.ylabel('End point')
plt.hist(path_end_point)
```

```
# Second kind of Randam Walk
step_number = 10000
step_set = [-1, 1]
origin = np.zeros((1,1))
step_shape = (step_number,1)
steps = np.random.choice(a=step_set, size=step_shape, p=[0.5, 0.5])

path = np.concatenate([origin, steps]).cumsum(0)
plt.title("Single Trajectory")
plt.xlabel('Step')
plt.ylabel('Location')
plt.plot(path)
```

```
path_end_point = []
for i in range(10000): 
    step_number = 10000 
    step_set = [-1, 1]
    origin = np.zeros((1,1))
    step_shape = (step_number,1)
    steps = np.random.choice(a=step_set, size=step_shape, p=[0.5, 0.5])
    path = np.concatenate([origin, steps]).cumsum(0)
    path_end_point.append(path[-1][0])

plt.title("End point of all Random walk")
plt.xlabel('Episode')
plt.ylabel('End point')
plt.hist(path_end_point)
```
